```python
import requests
from bs4 import BeautifulSoup
```

```python
# Target URL
url = 'https://www.prothomalo.com/entertainment/bollywood/50tjv1lck7'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
}
```

```python
# Send HTTP request
response = requests.get(url, headers=headers)
```

*   soup.find('div', class_='storyCard') → Finds a <div> with class "storyCard" (common container for article text).
*   find_all('p') → Extracts all paragraph tags <p> inside that container.

*   get_text(strip=True) → Gets clean text from each paragraph.
*   '\n'.join(...) → Joins all paragraphs together into a single text block separated by line breaks.


```python
# Check for successful response

# status_code == 200 means “OK” — the webpage was fetched successfully.
# If not 200 (for example 404 or 500), the script prints an error message

if response.status_code == 200:

  # Creates a BeautifulSoup object to parse the HTML content of the webpage.
  # "response.text" is the raw HTML code from the web page.
  #'html.parser' tells BeautifulSoup to use Python’s built-in HTML parser.

    soup = BeautifulSoup(response.text, 'html.parser')



    # Extract title
    title = soup.find('h1').get_text(strip=True)

    # Extract date
    date = soup.find('time')
    date_text = date.get_text(strip=True) if date else "No date found"

    # Extract main content
    article_body = soup.find('div', class_='storyCard')
    paragraphs = article_body.find_all('p') if article_body else []
    content = '\n'.join(p.get_text(strip=True) for p in paragraphs)

    # Print results
    print("Title:", title)
    print("Date:", date_text)
    print("Content:\n", content)

else:
    print("Failed to fetch the page. Status code:", response.status_code)
```
